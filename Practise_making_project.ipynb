```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv("matches.csv")
```

```python
match.head()
```

```python
delivery = pd.read_csv("deliveries.csv")
```

```python
delivery.head()
```

```python
match_score = delivery.groupby(["match_id","inning"])["total_runs"].sum().reset_index()
```

```python
match_score = match_score[match_score["inning"] == 1]
```

```python
match_score.head()
```

```python
match_df = match.merge(match_score[["match_id","total_runs"]],left_on = "id",right_on = "match_id")
```

```python
match_df.head()
```

```python
match_df["team1"].unique()
```

```python
# apply replacement in team name of Delhi Daredevils to Delhi Capitals and Deccan Chargers to Sunrisers Hyderabad

match_df["team1"] = match_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match_df["team2"] = match_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")

match_df["team1"] = match_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match_df["team2"] = match_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians',
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       'Delhi Capitals']
```

```python
# find the only selected teams from all teams of match

match_df = match_df[match_df["team1"].isin(teams)]
match_df = match_df[match_df["team2"].isin(teams)]
```

```python
match_df.shape
```

```python
# find only data from dl_applied column == 0 because it is without rain matchs. full form of dl is Duckworth lewis-stern

match_df = match_df[match_df["dl_applied"] == 0]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
# features extraction from the all datas 

match_df = match_df[["match_id","total_runs","winner","city"]]
```

```python
# merge the match_df with deliveries DataFrame
delivery_df = match_df.merge(delivery,on = "match_id")
```

```python
delivery_df.head()
```

```python
delivery_df = delivery_df[delivery_df["inning"] == 2]
```

```python
# make the new columns name of current_run
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
#delivery_df["current_score"] = delivery_df.groupby("match_id").cumsum()["total_runs_y"]
```

```python
delivery_df
```

```python
delivery_df["runs_left"] = delivery_df["total_runs_x"] - delivery_df["current_score"]
```

```python
delivery_df.head()
```

```python
delivery_df["ball_left"] = 126 - (delivery_df["over"]*6 + delivery_df["ball"])  
```

```python
delivery_df.head()
```

```python
delivery_df.columns
```

```python
match.columns
```

```python
match.head()
```

```python
# Ensure 'player_dismissed' column has no NaN values
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

# Map the values correctly to binary (0 or 1)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == "0" else 1)

# Convert the column to integer
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)

# Perform cumulative sum by match_id group
delivery_df['wickets'] = 10 - delivery_df.groupby('match_id')['player_dismissed'].cumsum()

# Display the first few rows
delivery_df.head(10)
```

```python
delivery_df.tail()
```

```python
# crr -> runs/overs
delivery_df["crr"] = (delivery_df["current_score"]*6)/(120 - delivery_df["ball_left"])
```

```python
delivery_df.head()
```

```python
#rrr -> 
delivery_df["rrr"] = (delivery_df["runs_left"]*6)/delivery_df["ball_left"]
```

```python
delivery_df.head()
```

```python
delivery_df.columns
```

```python
match["result"].value_counts()
```

```python
# make the new result column
def result(row):
    for i in range(len(delivery_df)):
        if row["winner"][i] == row["batting_team"][i]:
            return 1
        else :
            return 0
```

```python
def result(row):
    return 1 if row["winner"] == row["batting_team"] else 0
```

```python
delivery_df["result"] = delivery_df.apply(result,axis = 1)
```

```python
delivery_df.head()
```

```python
delivery_df.dropna(inplace = True)
```

```python
delivery_df.isnull().sum()
```

```python
final_df = delivery_df[["batting_team","bowling_team","city","runs_left","ball_left","wickets","total_runs_x","crr","rrr","result"]]
```

```python
final_df.head()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df = final_df[final_df["ball_left"] != 0]
```

```python
final_df.shape
```

```python
# feature selection 
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
# apply training and testing on dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([("trf",OneHotEncoder(sparse_output=False,drop = "first"),["batting_team","bowling_team","city"])],remainder = "passthrough",
                       force_int_remainder_cols=False )
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
pipe = Pipeline(steps = [("step1",trf),("step2",LogisticRegression(solver = "liblinear"))])
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = pipe.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```
